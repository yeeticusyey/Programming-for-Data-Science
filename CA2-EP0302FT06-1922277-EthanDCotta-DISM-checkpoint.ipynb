# Employed ICT Manpower and Vacant Job Positions (Line Graph)

```python
def linegraph():
    
    #Display dataframe information and raw data from dataset
    print(f"Overall inforamtion of Data 1 Dataframe ({filename1}): \n")
    print(f"{data1.info()}\n")
    print(f"Overall inforamtion of Data 2 Dataframe ({filename2}): \n")
    print(f"{data2.info()}\n")
    
    data1_list = []
    for x in np.split(data1, 2):
        data1_list.append(x)
        
    data1p1 = data1_list[0].style.set_table_attributes("style='display:inline'").set_caption('Dataset 1 (Part 1)')
    data1p2 = data1_list[1].style.set_table_attributes("style='display:inline'").set_caption('Dataset 1 (Part 2)')
    display_html(data1p1._repr_html_() + data1p2._repr_html_(), raw = True)
    data2p1 = data2.head(8).style.set_table_attributes("style='display:inline'").set_caption('Dataset 2 (Part 1)')
    data2p2 = data2.tail(7).style.set_table_attributes("style='display:inline'").set_caption('Dataset 2 (Part 2)')
    display_html(data2p1._repr_html_() + data2p2._repr_html_(), raw = True)
    
    #Extract rows and values from dataset
    data1_years = data1["year"]
    data1_manpower = data1["manpower"]
    #data1_growth = data1["growth"]
    data1_selected = pd.Series(data1[(data1["year"] >= 2002) & (data1["year"] <= 2017)]["manpower"]).reset_index(drop=True)
    
    data2_years = data2["year"]
    data2_jobs = data2["total_infocomm_jobs"]
    data_diff = data2_jobs - data1_selected
    
    #Text-based analysis
    print(f"Text-based analysis for {filename1} and {filename2}:\n")
    print(f"The year with the lowest number of employed manpower is during {data1['year'][data1['manpower'].idxmin()]} with {data1_manpower.min()} employed manpower")
    print(f"The year with the highest number of employed manpower is during {data1['year'][data1['manpower'].idxmax()]} with {data1_manpower.max()} employed manpower")
    print(f"\nTotal number of manpower employed over the years: {data1_manpower.sum()} people")
    print(f"Average number of manpower employed per the year: {int(data1_manpower.mean())} people")
    print(f"Median number of manpower employed: {int(data1_manpower.median())} people")
    
    print(f"\nThe year with the lowest number of vacant jobs is during {data2_years[pd.Index(data_diff).get_loc(data_diff.min())]} with {data_diff.min()} vacant job positions available")
    print(f"The year with the highest number of vacant jobs is during {data2_years[pd.Index(data_diff).get_loc(data_diff.max())]} with {data_diff.max()} vacant job positions available")
    print(f"\nTotal number of vacant job positions over the years: {data_diff.sum()} jobs")
    print(f"Average number of vacant job positions per the year: {int(data_diff.mean())} jobs")
    print(f"\nAdditonal statistical analysis for raw data: \n\n{data1['manpower'].describe()}\n\n{data2['total_infocomm_jobs'].describe()}")
    
    #Line Graph plot
    fig = plt.figure(figsize = (16, 8))
    plt.xlabel("Years", fontsize = 12)
    plt.ylabel("People/Manpower", fontsize = 12)
    plt.title("Line Graph of Total Employed ICT Manpower", fontsize = 16)
    plt.grid(axis = 'y', alpha = 0.75)
    plt.grid(axis = 'x', alpha = 0.75)
    plt.xticks(np.arange(data1_years.min(), data1_years.max() + 1, 2))
    plt.plot(data1_years, data1_manpower, 'b', label = "Total Employed Manpower")
    bf1 = plt.plot(data1_years, np.poly1d(np.polyfit(data1_years, data1_manpower, 1))(data1_years), 'b', alpha = 0.25, label = "Total Employed Manpower (Trend)")
    plt.plot(data2_years, data_diff, 'r', label = "Total Vacant Jobs")
    bf1 = plt.plot(data2_years, np.poly1d(np.polyfit(data2_years, data_diff, 1))(data2_years), 'r', alpha = 0.25, label = "Total Vacant Jobs (Trend)")
    legend = plt.legend(loc = 'lower right', bbox_to_anchor = (1, -0.3))
    plt.show()
    
    #Data Anaylsis and Conclusion
    percentage_manpower = float(((data1_manpower.max() - data1_manpower.min()) / data1_manpower.min()) * 100)
    percentage_vacant = float(((data_diff.max() - data_diff.min()) / data_diff.min()) * 100)
    print(f"According to the graph, it is evident that the general trend for each graph, Total Employed \nManpower (between {data1_years.min()} and {data1_years.max()}) and Total Vacant Jobs(between {data2_years.min()} and {data2_years.max()}) shows a steady \nincrease and slight increase respectively. There is an oddity during 2000 to 2001 of the graph that represents the total \nemployed manpower where it decreases from its original value. Additonally, there are multiple fluctuations in the graph \nreprsenting the total vacant jobs where the gradient is negative from 2005, 2008, 2010 and 2016.")
    print(f"\nThe percentage increase of Total Employed Manpower is {percentage_manpower:.2f}% (from {data1_manpower.min()} to {data1_manpower.max()})\nand the percentage increase of Total Vacant Jobs is {percentage_vacant:.2f}% (from {data_diff.min()} to {data_diff.max()}) ")
    print(f"\nThis is certainly due to the growing demand for professionals in the Infocomm Industry as Singapore progresses towards \ndeveloping into a 'SMART' nation. This conclusion can be formed because of the greater increase in values of the total \nemployed manpower and vacant jobs, as well as the positive gradient from each best fit line graph.")
    print(f"\nA line graph was implemented to depict the over trend for both variables. The second line graph that refers to the \nnumber of vacant jobs was derived from finding the difference between the total employed and vacanies minus the \nemployed manpower. They are appropriate for spotting the difference between multiple variables (2 in this case).\nThey also allow us to see the trend along with the best fit lines that displayed the rate of increase.")
linegraph()
```

# Total Infocomm Indusrty Revenue Generated (Bar Chart)

```python
def barchart():
    
    #Display dataframe information and raw data from dataset
    print(f"Overall inforamtion of Data 3 Dataframe ({filename3}): \n")
    print(f"{data3.info()}\n")
        
    data3p1 = data3.head(8).style.set_table_attributes("style='display:inline'").set_caption('Dataset 3 (Part 1)')
    data3p2 = data3.tail(7).style.set_table_attributes("style='display:inline'").set_caption('Dataset 3 (Part 2)')
    display_html(data3p1._repr_html_() + data3p2._repr_html_(), raw = True)
    
    #Extract rows and values from dataset
    data3_years = data3["year"]
    data3_revenue = data3["infocomm_industry_revenue"]
    
    #Text-based analysis
    print(f"\nText-based analysis for {filename3}:\n")
    print(f"The year with the lowest sum of revenue earned is during {data3['year'][data3['infocomm_industry_revenue'].idxmin()]} at ${float(data3_revenue.min())} (billion)")
    print(f"The year with the highest sum of revenue earned is during {data3['year'][data3['infocomm_industry_revenue'].idxmax()]} at ${float(data3_revenue.max())} (billion)")
    print(f"\nTotal amout of money earned over the years: ${float(data3_revenue.sum())} (billion)")
    print(f"Average amout of money earned per year: ${float(data3_revenue.mean()):.2f} (billion)")
    print(f"Median amout of money earned over the years: ${float(data3_revenue.median())} (billion)\n")
    print(f"Additonal statistical analysis for raw data: \n\n{data3['infocomm_industry_revenue'].describe()}\n")
    
    #Line Graph plot
    fig = plt.figure(figsize = (16, 8))
    plt.xlabel("Years", fontsize = 12)
    plt.ylabel("Revenue (Billion/$)", fontsize = 12)
    plt.title("Bar Graph of Total Infocomm Industry Revenue", fontsize = 16)
    plt.grid(axis = 'y', alpha = 0.75)
    plt.grid(axis = 'x', alpha = 0.75)
    plt.xticks(np.arange(data3_years.min(), data3_years.max() + 1, 2))
    plt.bar(data3_years, data3_revenue, color = (0.1, 0.8, 1.0), edgecolor = 'black', linewidth = 1.2, label = "Total Employed Manpower")
    plt.plot(data3_years, np.poly1d(np.polyfit(data3_years, data3_revenue, 1))(data3_years), 'b', label = "Total Employed Manpower (Trend)")
    legend = plt.legend(loc = 'lower right', bbox_to_anchor = (1, -0.2))
    plt.show()
    
    #Data Anaylsis and Conclusion
    percentage_revenue = float(((data3_revenue.max() - data3_revenue.min()) / data3_revenue.min()) * 100)
    print(f"\nWith reference to the bar chart, the overall trend from {data3_years.min()} to {data3_years.max()} shows a expotentially high increase of the total revenue \ngenerated from the infocomm industry. There is however, an anomaly during the year 2016 where the revenue dropped to lower than the previous value that is relative to it.The reason behind the decrease was on 1 October where Infocomm Media Development \nAuthority and GovTech are formed to take on a changing media landscape, resulting in the dissolution of the Media Development \nAuthority and Infocomm Development Authority of Singapore.")
    print(f"\nThe percentage increase of the Total Revenue Generated is {percentage_revenue:.2f}% (from ${data3_revenue.min()} billion to ${data3_revenue.max()} billion)")
    print(f"\nIn addition, the great increase in the total revenue generated from the infocomm industry was because of the induction of \nprofessionally trained and competent infocomm technology personnel. This can also be supported by Singapore's growing economy\nas it progresses towards a propsperous nation.")
    print(f"\nA bar graph was used to display the data visualisation because it clearly shows the total value it holds with each length of \nbar and allows for accurate comparison between each year in accordance to their respective numbers that are relative to \neach other. A best fit line was included to reinforce the conclusion that the overall trend is increasing, so as to support \nthis idea.") 
barchart()
```
