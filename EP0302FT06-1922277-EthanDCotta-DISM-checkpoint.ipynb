# Initialise Datasets and Python Libraries 

```python
#Import libraries for data visulisation and processing 

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

#Imported datafiles from https://data.gov.sg/search?groups=society

#https://data.gov.sg/dataset/iras-collection-by-tax-type-annual
filename1 = "ca1_files\iras-collection-by-tax-type-annual\iras-collection-by-tax-type.csv"
data1 = np.genfromtxt(filename1, dtype = ['i2', 'U50', 'i8'], 
                         delimiter = ',', names = True)

#https://data.gov.sg/dataset/annual-value-and-property-tax-by-property-type-annual
filename2 = "ca1_files/annual-value-and-property-tax-by-property-type-annual/median-annual-value-and-property-tax-by-type-of-private-residential-property.csv"
data2 = np.genfromtxt(filename2, dtype = ['i2', 'U50', 'i8', 'i8', 'i8'], 
                         delimiter = ',', names = True)

#https://data.gov.sg/dataset/annual-value-and-property-tax-by-property-type-annual
filename3 = "ca1_files/annual-value-and-property-tax-by-property-type-annual/annual-value-and-property-tax-by-property-type.csv"
data3 = np.genfromtxt(filename3, dtype = ['i2', 'U50', 'i8', 'i8', 'i8', 'i8'], 
                         delimiter = ',', names = True)

#https://data.gov.sg/dataset/private-residential-property-price-index-by-type-of-property
filename4 = "ca1_files/private-residential-property-price-index-by-type-of-property/private-residential-property-price-index-base-quarter-2009-q1-100.csv"
data4 = np.genfromtxt(filename4, dtype = ['U7', 'U50', 'float'], 
                         delimiter = ',', names = True)

#https://data.gov.sg/dataset/completed-private-residential-units-sold-in-the-quarter-quarterly
filename5 = "ca1_files/completed-private-residential-units-sold-in-the-quarter-quarterly\completed-private-residential-units-sold-in-the-quarter-quarterly.csv"
data5 = np.genfromtxt(filename5, dtype = ['U7', 'U50', 'i8'], 
                         delimiter = ',', names = True)

#https://data.gov.sg/dataset/annual-value-and-property-tax-by-property-type-annual
filename6 = "ca1_files/annual-value-and-property-tax-by-property-type-annual\median-annual-value-and-property-tax-by-type-of-hdb.csv"
data6 = np.genfromtxt(filename6, dtype = ['i2', 'U50', 'i8', 'i8', 'i8'], 
                         delimiter = ',', names = True)

blank = ""
```

# Overall Trend of Property Tax Collected (Bar Chart)

```python
#Bar Chart for dataset 1 ("iras-collection-by-tax-type.csv")

def barchart_plot():
    
    #Extract rows and values from dataset
    data1_years = np.unique(data1['financial_year'])
    data1_proptax = data1[data1['tax_type'] == 'Property Tax']['tax_collected']
    
    #Print statements for raw data 
    print(f"Data analysis for {filename1}\nof the total amount of money collected from property taxes over {len(data1_years)} years from {data1_years.min()} to {data1_years.max()}")
    print(f"\nThe year with the lowest amount of tax collected is {data1[data1['tax_collected'] == data1_proptax.min()]['financial_year'][0]} at ${data1_proptax.min()}")
    print(f"The year with the highest amount of tax collected is {data1[data1['tax_collected'] == data1_proptax.max()]['financial_year'][0]} at ${data1_proptax.max()}")
    print(f"\nTotal sum of tax collected: ${data1_proptax.sum()}")
    print(f"Average amount of tax collected per year: ${data1_proptax.mean():.2f}")
    print(f"Median of tax collected: ${np.median(data1_proptax):.2f}")
    print(f"\n*Raw Data (table)*\n\n   Year   : Tax Collected ($)\n{blank:-^30}")
    for x in range(len(data1_proptax)):
        print(f"Year {data1_years[x]} :    ${data1_proptax[x]}")
    
    #Bar Chart plot
    fig = plt.figure(figsize = (12, 8))
    plt.title("Bar Chart of Total Property Tax Collected", fontsize = 16)
    plt.xlabel("Years", fontsize = 12)
    plt.ylabel("Dollars ($/million)", fontsize = 12)
    plt.xlim((data1_years.min() - 1), (data1_years.max() + 1))
    plt.xticks(np.arange(data1_years.min(), data1_years.max() + 1, 2))
    #plt.yticks(np.arange(round(data1_proptax.min(), -6), round(data1_proptax.max(), -6), 1000000))
    plt.grid(axis = 'y', alpha = 0.75)
    bar_chart = plt.bar(data1_years, data1_proptax, color = (0.1, 0.8, 1.0), edgecolor = 'black', linewidth = 1.2)
    plt.show()
    
    #Data anaylsis and conclusion
    percent_diff = float(((data1_proptax.max() - data1_proptax.min()) / data1_proptax.min()) * 100)
    print(f"According to the bar chart, there is an overall increase in amount of property collected from {data1_years.min()} to {data1_years.max()}.")
    print(f"There was however, an anomaly in the increasing trend during 2009, where the total amount of property tax \ncollected decreased significantly.")
    print(f"\nThe percentage increase of property taxes collected from {data1_years.min()} to {data1_years.max()} is {percent_diff:.2f}% (from ${data1_proptax.min()} to ${data1_proptax.max()})")
    print(f"\nA bar chart was used for this data analysis because the apex and length of each bar would easily depict the\namount of tax collected relative to the others. Therefore, the trend would be identified properly.")
    
barchart_plot()
```

# Non-landed and Landed Property Tax Collection (Line Graph)

```python
#Line graph for dataset 2 ("median-annual-value-and-property-tax-by-type-of-private-residential-property.csv")

def linegraph_plot():
    
    #Extract rows and values from dataset
    data2_years = np.unique(data2['financial_year'])
    data2_nonlanded = data2[data2['type_of_private_residential'] == 'Non-landed (includes Executive Condominiums)']
    data2_landed = data2[data2['type_of_private_residential'] == 'Landed']
    data2_nonlanded_tax = data2_nonlanded['property_tax_collection']
    data2_landed_tax = data2_landed['property_tax_collection']

    #Print statements for raw data
    print(f"Data analysis of {filename2} concerning the total amount of money collected from non-landed and landed property taxes \nover {len(data2_years)} years from {data2_years.min()} to {data2_years.max()}")
    print(f"\nThe year with the lowest amount of non-landed tax collected is {data2_nonlanded[data2_nonlanded['property_tax_collection'] == data2_nonlanded_tax.min()]['financial_year'][0]} at ${data2_nonlanded_tax.min()}")
    print(f"The year with the highest amount of non-landed tax collected is {data2_nonlanded[data2_nonlanded['property_tax_collection'] == data2_nonlanded_tax.max()]['financial_year'][0]} at ${data2_nonlanded_tax.max()}")
    print(f"The year with the lowest amount of landed tax collected is {data2_landed[data2_landed['property_tax_collection'] == data2_landed_tax.min()]['financial_year'][0]} at ${data2_landed_tax.min()}")
    print(f"The year with the highest amount of landed tax collected is {data2_landed[data2_landed['property_tax_collection'] == data2_landed_tax.max()]['financial_year'][0]} at ${data2_landed_tax.max()}")
    print(f"\nTotal sum of non-landed tax collected: ${data2_nonlanded_tax.sum()}")
    print(f"Average amount of non-landed tax collected per year: ${data2_nonlanded_tax.mean():.2f}")
    print(f"Median of non-landed tax collected: ${np.median(data2_nonlanded_tax):.2f}")
    print(f"\nTotal sum of landed tax collected: ${data2_landed_tax.sum()}")
    print(f"Average amount of landed tax collected per year: ${data2_landed_tax.mean():.2f}")
    print(f"Median of landed tax collected: ${np.median(data2_landed_tax):.2f}")
    print(f"\n*Raw Data (table)*\n\n   Year   : Tax Collected ($/Non-landed) | Tax Collected ($/Landed) | Difference ($)\n{blank:-^84}")
    for x in range(len(data2_nonlanded_tax)):
        print(f"Year {data2_years[x]} :{blank: ^11}${data2_nonlanded_tax[x]}{blank: ^12}|{blank: ^10}${data2_landed_tax[x]}{blank: ^9}|{blank: ^4}${data2_nonlanded_tax[x] - data2_landed_tax[x]}")
        
    #Line Graph plot
    fig = plt.figure(figsize = (12, 8))
    plt.title("Line Graph of Landed and Non-landed Property Tax Collected", fontsize = 16)
    plt.grid(axis = 'y', alpha = 0.75)
    plt.grid(axis = 'x', alpha = 0.75)
    plt.xlabel("Years", fontsize = 12)
    plt.ylabel("Dollars ($)", fontsize = 12)
    line1 = plt.plot(data2_years, data2_nonlanded_tax, 'b', label = "Non-landed (includes Executive Condominiums)")
    line2 = plt.plot(data2_years, data2_landed_tax, 'r', label = "Landed")
    point1 = plt.plot(data2_years, data2_nonlanded_tax, 'bo')
    point2 = plt.plot(data2_years, data2_landed_tax, 'ro')
    bf1 = plt.plot(data2_years, np.poly1d(np.polyfit(data2_years, data2_nonlanded_tax, 1))(data2_years), 'b', alpha = 0.25)
    bf2 = plt.plot(data2_years, np.poly1d(np.polyfit(data2_years, data2_landed_tax, 1))(data2_years), 'r', alpha = 0.25)
    plt.vlines(x = data2_years, ymin = data2_landed_tax, ymax = data2_nonlanded_tax, colors = 'k', ls = '--', lw = 1, label = 'Difference')
    legend = plt.legend(loc = 'lower right')
    plt.show()
    
    #Data anaylsis and conclusion
    percent_diff_nonlanded = float((data2_nonlanded_tax.max() - data2_nonlanded_tax.min()) / data2_nonlanded_tax.min() * 100)
    percent_diff_landed = float((data2_landed_tax.max() - data2_landed_tax.min()) / data2_landed_tax.min() * 100)
    print(f"According to the line graphs of each type of private property, both show a gradual increase in amount of tax collected, as well as the difference between the 2. The increase of non-landed property tax collection is greater than that of landed ones since the gradient for that line is slightly steeper.")
    print(f"\nThe non-landed line graph shows an increase in property tax collection from 2010 to 2015 and then gradually decreases from 2015 to 2019. Overall there is still a rising trend.")
    print(f"\nAs for the landed graph, it shares the exact same characteristic as the non-landed one, except there is an oddity during 2012 \nwhere it extremely slightly decreased.")
    print(f"\nThe percentage increase of non-landed property taxes collected from {data2_years.min()} to {data2_years.max()} is {percent_diff_nonlanded:.2f}% (from ${data2_nonlanded_tax.min()} to ${data2_nonlanded_tax.max()}. On the \nother hand, The percentage increase of landed property taxes collected from {data2_years.min()} to {data2_years.max()} is {percent_diff_landed:.2f}% (from ${data2_landed_tax.min()} to ${data2_landed_tax.max()})")
    print(f"\nLine graphs were used for this analysis because they are appropriate for spotting the difference between multiple variables (2 in this case). They also allow us to see the trend along with the best fit lines that displayed the rate of increase.")
                                   
linegraph_plot()
```

# Property Tax Collected by Property Type (Pie Chart)

```python
#Pie Chart for dataset 3 ("median-annual-value-and-property-tax-by-type-of-private-residential-property.csv")

def piechart_plot():
    
    #Extract rows and values from dataset
    data3_years = np.unique(data3['financial_year'])
    data3_proptype = np.unique(data3['property_type'])
    
    data3_2002 = data3[(data3['financial_year'] == 2002)]
    chart1 = data3_2002[np.isin(data3_2002['property_type'],['Commercial', 'HDB', 'Industrial', 'Others', 'Private Residential'])]['property_tax_collection']
    
    data3_2019 = data3[(data3['financial_year'] == 2019)]
    chart2 = data3_2019[np.isin(data3_2019['property_type'],['Commercial', 'HDB', 'Industrial', 'Others', 'Private Residential'])]['property_tax_collection']
    
    #Print statements for raw data
    print(f"Data analysis of {filename3} between the total amount of tax collected from {len(data3_proptype)} different property types in {data3_years.min()} and {data3_years.max()}")
    print(f"\nTotal sum of tax collected in {data3_years.min()} by different property type: ${chart1.sum()}")
    print(f"Average amount of tax collected in {data3_years.min()} by different property type: ${chart1.mean():.2f}")
    print(f"Median of tax collected in {data3_years.min()} by different property type: ${np.median(chart1):.2f}")
    print(f"\nTotal sum of tax collected in {data3_years.max()} by different property type: ${chart2.sum()}")
    print(f"Average amount of tax collected in {data3_years.max()} by different property type: ${chart2.mean():.2f}")
    print(f"Median of tax collected in {data3_years.max()} by different property type: ${np.median(chart2):.2f}")
    print(f"\n*Raw Data (table)*\n\n    Property Type   : Total tax collected in 2002 ($) |   Total tax collected in 2019 ($)   | Difference ($)\n{blank:-^107}")
    for x in range(len(data3_proptype)):
        print(f"{data3_proptype[x]:<19s} :{blank: ^13}${str(chart1[x]):<19s}|{blank: ^15}${str(chart2[x]):<21s}|{blank: ^4}${chart2[x] - chart1[x]}")
    
    #Pie Chart plot
    fig, ax = plt.subplots(1, 2, figsize = (15, 10))
    colors = ['lightblue', 'red', 'purple', 'blue', 'cyan']
    labels = ['HDB', 'Private Residential', 'Commercial', 'Industrial', 'Others']
    explode = (0, 0, 0.1, 0.1, 0)
    ax[0].pie(chart1, labels = labels, colors = colors, autopct='%1.2f%%', shadow=True)
    ax[1].pie(chart2, labels = labels, colors = colors, autopct='%1.2f%%', shadow=True, explode = explode)
    plt.title("yeet")
    ax[0].set_title("Total tax collected in 2002", fontsize = 14)
    ax[1].set_title("Total tax collected in 2019", fontsize = 14)
    plt.show()
    
    #Data anaylsis and conclusion
    print(f"According to the 2 pie charts, there was a substantial in increase in proportion of tax collected from Commerical and \nIndustrial properties and a decrease for Private, HDB and Others from 2002 to 2019. ")
    print(f"\nPie Charts were used to display the differences since we are now working with 5 variables (property types) and it gives us a \nclearer way to compare them. A line graph would be too messy and disorganised that may lead to false interpretation of the \ndata. Since we are working with mulitple components, it is more convenient to decipher the percentage difference and portions.")
    
piechart_plot()
```

# Comparison of Residential Property Price Index (Histogram & Scatter Plot)

```python
#Histogram for dataset 4 ("private-residential-property-price-index-base-quarter-2009-q1-100.csv")

def histogram_plot():
    
    #Extract rows and values from dataset
    data4_quarters = data4[data4['quarter'] > '2002']
    data4_years = data4_quarters['quarter']
    data4_values = data4_quarters['value']
    
    #Print statements for raw data
    print(f"Data analysis of {filename4} for the residential property price index of {len(data4_quarters)} quarters from {data4_years[0]} to {data4_years[len(data4_quarters) - 1]}")
    print(f"\nLowest residential property price index property: {data4_values.min()}")
    print(f"Highest residential property price index property: {data4_values.max()}")
    print(f"\nAverage residential property price index property: {data4_values.mean():.2f}")
    print(f"Median residential property price index property: {np.median(data4_values):.2f}")
    print(f"\n*Raw data (list)*\n")
    for x in data4_values:
        print(x, end = "  ")
    
    #Histogram plot
    fig = plt.figure(figsize = (12, 8))
    plt.title("Histogram of Residential Property Price Index", fontsize = 16)
    plt.grid(axis = 'y', alpha = 0.5)
    plt.xlabel("Value (Price Index)", fontsize = 12)
    plt.ylabel("Frequency", fontsize = 12)
    plt.hist(data4_values, color = (0.1, 0.8, 1.0), edgecolor = 'black', linewidth = 1.2)
    plt.show()
    
    #Data anaylsis and conclusion
    print(f"According to the histogram, there are greater frequencies of the residential property price index at the extreme ends of the \nrange (approximatelty 80 and 150). Between the price index of 90 to 140, there is a steady increase of frequencies. ")
    print(f"\nThere is an anomaly between the price index from approximately 100 to 110 where it shows a decrease instead of increase.")
    print(f"\nA histogram was used to measure the data since we are dealing with a large sample size, so it would be more appropriate to \nconsolidate the information and present it through a histogram.")
histogram_plot()
```

```python
#Scatter Plot for dataset 4 & 5 ("private-residential-property-price-index-base-quarter-2009-q1-100.csv, 
#                                completed-private-residential-units-sold-in-the-quarter-quarterly.csv")

def scatter_plot():
    
    #Extract rows and values from dataset
    data4_quarters = data4[data4['quarter'] > '2002']
    data4_values = data4_quarters['value']
    data4_years = data4_quarters['quarter']
    
    data5_quarters = data5[data5['quarter'] > '2002']
    data5_values = data5_quarters['units']
    
    #Print statements for raw data
    print(f"Data analysis of {filename4} and {filename5} for the residential property price index of {len(data4_quarters)} quarters from {data4_years[0]} \nto {data4_years[len(data4_quarters) - 1]}")
    print(f"\nLowest residential property price index property: {data4_values.min()}")
    print(f"Highest residential property price index property: {data4_values.max()}")
    print(f"\nAverage residential property price index property: {data4_values.mean():.2f}")
    print(f"Median residential property price index property: {np.median(data4_values):.2f}")
    print(f"\nTotal sum of number for units sold: {data5_values.sum()}")
    print(f"Lowest number for units sold: {data5_values.min()}")
    print(f"Highest number for units sold: {data5_values.max()}")
    print(f"\nAverage number for units sold: {data5_values.mean():.2f}")
    print(f"Median of number for units sold: {np.median(data5_values):.2f}")
    
    print(f"\n*Raw data for price index (list)*\n")
    for x in data4_values:
        print(x, end = "  ")
        
    print(f"\n\n*Raw data for number of units sold (list)*\n")
    for x in data5_values:
        print(x, end = "  ")
    
    #Scatter Graph plot
    fig = plt.figure(figsize = (12, 8))
    plt.xlabel("Units Sold", fontsize = 12)
    plt.ylabel("Residential Property Price Index", fontsize = 12)
    plt.title('Scatter Plot Graph of Price Index Against Number of Units Sold', fontsize = 16)
    plt.grid(axis = 'x', alpha = 0.5)
    plt.grid(axis = 'y', alpha = 0.5)
    plt.scatter(data5_values, data4_values, label = 'Scatter Plot Points')
    plt.plot(data5_values, np.poly1d(np.polyfit(data5_values, data4_values, 1))(data5_values), 'r', label = 'Best Fit Line')
    legend = plt.legend(loc = 'upper right')
    plt.show()
    
    #Data anaylsis and conclusion
    print(f"According to the scatter plot graph and best fit line, it is apparent that there are lesser units sold where the residential property price index is greater and more when it is lower. This is due to a larger portion of the population in singapore falling into the lower and medium socio-economic status.")
    print(f"\nThe downward gradient of the best fit line symbolises the aforementioned. ")
    print(f"\nA scatter plot grpah was used to compare the price index against the number of units sold because it would clearly show the relation between the 2 variables and display the overall trend. ")
scatter_plot()
```

# Median Annual Property Tax of HDB Type (Box Plot)

```python
#Histogram for dataset 6 ("median-annual-value-and-property-tax-by-type-of-hdb.csv")

def box_plot():
    
    #Extract rows and values from dataset
    data6_years = np.unique(data6['financial_year'])
    data6_hdbtype = np.unique(data6['type_of_hdb'])
    
    rooms_1to2 = data6[data6['type_of_hdb'] == '1 or 2 Room']['median_annual_value']
    rooms_3 = data6[data6['type_of_hdb'] == '3 Room']['median_annual_value']
    rooms_4 = data6[data6['type_of_hdb'] == '4 Room']['median_annual_value']
    rooms_5 = data6[data6['type_of_hdb'] == '5 Room']['median_annual_value']
    exec_oth = data6[data6['type_of_hdb'] == 'Executive & Others']['median_annual_value']
    med_values = np.array([rooms_1to2, rooms_3, rooms_4, rooms_5, exec_oth])
    
    #Print statements for raw data
    print(f"Data analysis of {filename6} for the median annual property tax by HDB type of {len(data6_years)} years from {data6_years.min()} to {data6_years.max()}")
    print(f"\n*Raw Data (table)*\n\n   Year   : 1-2 Rooms | 3 Rooms |  4 Rooms  |  5 Rooms  | Executive and Others\n{blank:-^78}")
    for x in range(len(rooms_1to2)):
        print(f"Year {data6_years[x]} :   ${rooms_1to2[x]}   |  ${rooms_3[x]}  |   ${str(rooms_4[x]):<5s}  |   ${str(rooms_5[x]):<5s}  |       ${exec_oth[x]}")
    
    #Box Plot 
    fig = plt.figure(figsize = (12, 8))
    plt.xlabel("Type of HDB Unit", fontsize = 12)
    plt.ylabel("Median Annual Property Tax($)", fontsize = 12)
    plt.grid(axis = 'y', alpha = 0.5)
    plt.title('Box Plot of Median Annual Property Tax of HDB Type', fontsize = 16)
    plt.boxplot(med_values.transpose(), labels = data6_hdbtype)
    plt.show()

    #Data anaylsis and conclusion
    print(f"According to the box plot, it is shown that the greater the number of rooms or standard of HDB type, the greater the median \nannual property tax. Each category has an outlier that is the minimum value at year 2010. ")
    print(f"\nA box plot was used to depict the data as it presents a deeper understanding for the range of median annual property taxes for each HDB type. ")
    
    
box_plot()
```
